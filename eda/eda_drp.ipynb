### Organizational structure

```python
org_counts = df['organization'].value_counts()
org_counts
```

```python
top_orgs = org_counts.head(20)

plt.figure()
plt.bar(top_orgs.index, top_orgs.values)
plt.xticks(rotation=90)
plt.xlabel('Organization')
plt.ylabel('Number of datasets')
plt.title('Distribution of datasets by organization (Top 20)')
plt.show()
```

```python
(org_counts == 1).mean()
```

65 different organisations, with 26% being represented by only one example

```python
agency_counts = df['agency'].value_counts()
agency_counts
```

```python
top_agencies = agency_counts.head(20)

plt.figure()
plt.bar(top_agencies.index, top_agencies.values)
plt.xticks(rotation=90)
plt.xlabel('Agency')
plt.ylabel('Number of datasets')
plt.title('Distribution of datasets by agency (Top 20)')
plt.show()
```

```python
(agency_counts == 1).mean()
```

34 different agencies, with 14.7% being represented by only one example

### Rescue efforts

```python
df['maintainer'].value_counts()
```

Most of the data in the dataset is maintained now by Data Rescue Project and DataLumos (as expected).

```python
dc_created = pd.to_datetime(df['dc_created'], errors='coerce')

created_month = dc_created.dt.to_period("M")
created_per_month = (created_month.value_counts().sort_index())
created_per_month.index = created_per_month.index.to_timestamp()

plt.figure()
plt.plot(created_per_month.index, created_per_month.values)
plt.xlabel("Month")
plt.ylabel("Number of datasets created")
plt.title("Datasets created per month (dc_created)")
plt.show()

created_since_2025 = created_per_month[created_per_month.index >= pd.Timestamp("2025-01-01")]

plt.figure()
plt.plot(created_since_2025.index, created_since_2025.values)
plt.xlabel("Month")
plt.ylabel("Number of datasets created")
plt.title("Datasets created per month since January 2025")
plt.show()
```
