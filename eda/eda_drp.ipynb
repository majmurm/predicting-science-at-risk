### Rescue efforts

```python
df['maintainer'].value_counts()
```

Most of the data in the dataset is maintained now by Data Rescue Project and DataLumos (as expected).

```python
dc_created = pd.to_datetime(df['dc_created_y'], errors='coerce')

created_month = dc_created.dt.to_period("M")
created_per_month = (created_month.value_counts().sort_index())
created_per_month.index = created_per_month.index.to_timestamp()

plt.figure()
plt.plot(created_per_month.index, created_per_month.values)
plt.xlabel("Month")
plt.ylabel("Number of datasets created")
plt.title("Datasets created per month (dc_created_y)")
plt.show()

created_since_2025 = created_per_month[created_per_month.index >= pd.Timestamp("2025-01-01")]

plt.figure()
plt.plot(created_since_2025.index, created_since_2025.values)
plt.xlabel("Month")
plt.ylabel("Number of datasets created")
plt.title("Datasets created per month since January 2025")
plt.show()
```
